# testing single neuron for forward and backward propogation
(activation function : sigmoid, loss function: binary cross entropy)

```
import torch
```

```
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```
## binary cross entropy loss func.
def binary_cross_entropy_loss(pred, targt):
  epsilon = 1e-8
  pred = torch.clamp(pred, epsilon , 1- epsilon)
  return -(targt * torch.log(pred) + (1-targt) * torch.log(1-pred))
```

```
z = w*x +b
pred = torch.sigmoid(z)
loss = binary_cross_entropy_loss(pred, y)
```

```
z
```

```
loss
```

```
## backward propogation
loss.backward()
print(w.grad)
print(b.grad)
```
